#### **5.2.2 Data Cleaning**
<p align = "justify">Our next step is to carry out a data cleaning process. This involves dealing with missing values and standardizing the column names of every dataset for consistency. </p>

```python
import pandas as pd

# Get CSV file
df = pd.read_csv("/content/drive/MyDrive/dataset.csv")

num_columns1 = df.shape[0]
print("Before modification:", num_columns1)

# Handling missing values
df.dropna(inplace=True)

# Handling missing values
# For example, suppose you want to handle outliers in a numerical column named 'Age at enrollment'
# Using the IQR (interquartile range) method
Q1 = df['Age at enrollment'].quantile(0.25)
Q3 = df['Age at enrollment'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Age at enrollment'] >= lower_bound) & (df['Age at enrollment'] <= upper_bound)]

num_columns2 = df.shape[0]
print("After modification:", num_columns2)

# handling data inconsistencies
# For example, suppose you want to unify the processing of a target column named 'Target' and convert it to a unified representation
df['Target'] = df['Target'].map({'dropout': 'Dropout', 'enrolled': 'Enrolled', 'graduate':'Graduate'})
```

Based on the representation structure of the target, we adopt the Label Encoding method. Convert 'Gradate', 'Dropout', and 'Enrolled' to 0, 1, 2.</p>
<p align = "justify">As for what is 0, what is 1, and what is 2, it is based on the order in which they appear in the data columns. For example, if 'Gradate' appears first, 'Gradate' will be assigned 0; If 'Dropout' appears again, it will be assigned a value of 1, and so on.</p>

## **6.0 Modelling**

### **6.1 Logistic Regression**

Logistic regression is particularly well-suited for the binary classification task of predicting student dropouts in our project. This model estimates the probability of a binary response based on one or more predictor variables, making it ideal for scenarios where you want to understand the impact of various factors such as attendance, grades, and socio-economic status on student outcomes. The output of logistic regression is a logistic curve, which means the probability of dropping out is modeled as a function of the predictors. This allows for easy interpretation of how each feature influences the likelihood of a student dropping out or succeeding academically. Moreover, logistic regression can handle both numerical and categorical data, which is advantageous when dealing with diverse datasets. It also provides useful model diagnostics and metrics such as odds ratios that can be communicated to educational stakeholders to inform interventions.

### **6.2 Decision Trees and Random Forests**

Decision trees provide a clear visual representation of decision paths, which can be invaluable in understanding complex decision-making processes behind student success. Random Forests, an ensemble of decision trees, enhance predictive accuracy and robustness against overfitting by averaging multiple trees, making it highly effective for handling diverse datasets.

### **6.3 Gradient Boosting Machines (GBM)**

GBM is a powerful ensemble technique that builds trees sequentially, with each new tree correcting errors made by previous trees. This approach can be very effective in handling unbalanced datasets, which is common in educational data where dropout rates might be low relative to success rates.

### **6.4 Neural Networks**

These models are capable of modeling complex nonlinear relationships and interactions among multiple variables. For educational datasets, which often contain a wide range of influencing factors from socio-economic to academic performance metrics, neural networks can capture these intricate patterns effectively, potentially leading to higher predictive performance.
